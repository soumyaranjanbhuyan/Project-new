# AI Financial Market Data for Gemini(Google), ChatGPT(OpenAI), Llama(Meta)

#### This dataset provides a synthetic, daily record of financial market activities related to companies involved in Artificial Intelligence (AI). There are key financial metrics and events that could influence a company's stock performance like launch of Llama by Meta, launch of GPT by OpenAI, launch of Gemini by Google etc. Here, we have the data about how much amount the companies are spending on R & D of their AI's Products & Services, and how much revenue these companies are generating. The data is from January 1, 2015, to December 31, 2024, and includes information for various companies : OpenAI, Google and Meta.

#### 
The dataset captures company-level statistics on R\&D spending, AI-driven revenue, and corresponding stock price impacts.

1. R\&D Spending (USD Mn)

   * Companies, on average, invest around 65 million USD in R\&D, but the distribution is highly uneven.
   * The wide standard deviation (47.9) and the large gap between the 25th percentile (8.64) and the 75th percentile (99.60) suggest that while some firms invest minimally, others allocate significant budgets.
   * This indicates a polarized landscape—AI development is capital intensive for leading firms, while smaller players remain conservative.

2. AI Revenue (USD Mn)

   * Average AI revenue is 44 million USD, with high variability (std dev \~41.6).
   * The 25th percentile at 3.61 vs. 75th percentile at 71.68 shows that only a minority of firms generate substantial AI revenue.
   * Negative minimum (-0.55) may reflect accounting adjustments or loss entries.
   * Overall, AI adoption is still maturing—many companies earn little, but a few top firms dominate AI-driven income.

3. Stock Impact (%)

   * Stock movements tied to AI-related factors average near zero (0.025%), indicating that AI announcements typically don’t move markets for most companies.
   * The distribution is skewed: median = 0%, but extreme cases exist, with some firms experiencing up to +18.5% jumps and others losing up to -3%.
   * This suggests AI-driven stock reactions are event-specific, not uniform across firms.

Overall Insights:

* AI is a high-risk, high-reward domain—both R\&D spending and AI revenue are concentrated in a few firms.
* For most companies, AI doesn’t significantly influence stock prices yet, but for leaders, AI initiatives can trigger outsized market responses.
* This pattern highlights the growing but uneven economic and financial significance of AI investments.

#### How much amount the companies spent on R & D ?

```python
# "Company's spending for R & D in $Bn"
RD = df.groupby('Company')['R&D_Spending_USD_Mn'].sum()/1000
RD
```

```python
# Draw a Bar Plot to show the amount spent on R & D by the companies
plt.figure(figsize=(12, 8))
plt.bar(RD.index, RD.values, color = ['Orange', 'gold', 'yellow'], width=0.3)

plt.title( "R&D Spending by the companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")

plt.show()
```

#### Revenue Earned by the companies

```python
# "Company's AI_Revenue_USD_Bn "
rev = df.groupby('Company')['AI_Revenue_USD_Mn'].sum()/1000
rev
```

```python
# Draw a Bar Plot to show the Revenues of the companies
plt.figure(figsize=(12, 8))
plt.bar(rev.index, rev.values, color = ['Orange', 'gold', 'yellow'], width = 0.3)

plt.title( "Revenue earned by the companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")


plt.show()
```

```python
# Bar plots to show R & D expenditure & revenue of the companies

plt.figure(figsize = (10,4))

plt.subplot(1,2,1)

plt.bar(RD.index, RD.values, color = ['Orange', 'gold', 'yellow'],width=0.3)
plt.title("Revenue earned by the companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")

plt.subplot(1,2,2)

plt.bar(rev.index, rev.values, color = ['Orange', 'gold', 'yellow'], width = 0.3)

plt.title( "Revenue earned by the companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")
plt.show()

```
